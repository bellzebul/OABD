# 1. Stackplot

```python
years = [2005, 2006, 2009, 2014, 2016, 2019]
```

```python
data_years = data.loc[data['Starting Year'].isin(years)]
pd.set_option('display.max_rows', None)
```

```python
def get_amount_startaps(city: str, year: int):
    curr_city = data_years.loc[(data_years['City'] == city) & (data_years['Starting Year'] == year)]
    return len(curr_city)
```

```python
startaps_by_sity = {
    'Bengaluru': [get_amount_startaps('Bengaluru', i) for i in years],
    'Mumbai': [get_amount_startaps('Mumbai', i) for i in years],
    'Gurgaon': [get_amount_startaps('Gurgaon', i) for i in years],
    'Ahmedabad': [get_amount_startaps('Ahmedabad', i) for i in years],
    'New Delhi': [get_amount_startaps('New Delhi', i) for i in years],
}
```

```python
startaps_by_sity
```

```python
fig = plt.figure(figsize=(12, 5))

plt.stackplot(years, startaps_by_sity.values(), labels=startaps_by_sity.keys())
plt.xticks(years)
plt.legend(loc='upper left')
plt.title('Startups amount')
plt.xlabel('Year')
plt.ylabel('Number of startups')
plt.xticks(years)
```

# 2. Bar of pie

```python
from matplotlib.patches import ConnectionPatch
```

```python
def get_amount(industry):
    return list(data['Industries'].str.contains(industry)).count(True)
```

```python
def norm(data):
    return [round(i/sum(data), 2) for i in data]
```

```python
def get_years(industry):
    places = []
    ind_list = data['Industries'].str.contains(industry)
    for i in range(len(ind_list)):
        if ind_list[i] == True:  
            places.append(i)
            
    years_ranges = [0, 0, 0]
    for year in list(data.loc[places]['Starting Year']):
        if year < 2010:
            years_ranges[0]+=1
        elif 2010 <= year < 2015:
            years_ranges[1]+=1
        elif year >= 2015:
            years_ranges[2]+=1
    return years_ranges
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0.05)

industries = ['Apps', 'Lifestyle', 'Financial Services', 'Education']
overall_ratios = norm([get_amount(industry) for industry in industries])

angle = -180 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, autopct='%1.1f%%', startangle=angle,
                     labels=industries, explode = [0.1, 0, 0, 0], shadow=True)

ages = ['...2009', '2010...2015', '2016...']
age_ratios = norm(get_years('Apps'))

bottom = 1
width = .2

for j, (height, label) in enumerate(reversed([*zip(age_ratios, ages)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='C0', label=label,
                 alpha=0.1 + 0.25 * j)
    ax2.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')
    
ax2.set_title('Starting years')
ax2.legend()
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)

theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(age_ratios)

x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(-width / 2, bar_height), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(4)
ax2.add_artist(con)


x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(-width / 2, 0), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(4)
```
